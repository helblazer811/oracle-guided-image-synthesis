```python
import matplotlib.pyplot as plt
from dataset import MetadataDataset
import numpy as np
```

How are each of the morpho mnist metadata features distributed for each versus the entire range?

```python
# define a histogram plot
def histogram_plot(digit):
    feature_names = ["area", "length", "thickness", "slant", "width", "height"]
    dataset = MetadataDataset(which_digits=[digit], apply_transform=True)
    # plot the histogram
    num_examples = 500
    num_features = 6
    fig, axs = plt.subplots(1, num_features, figsize=(30, 4))
    metadata = []
    for example in range(num_examples):
        metadata.append(dataset[example])
    metadata = np.array(metadata)
    for feature_index in range(num_features):
        axs[feature_index].hist(metadata[:, feature_index], range=(0, 1))
        axs[feature_index].set_title(feature_names[feature_index])
    plt.show()
```

```python
#For each digit go through and plot histograms for each figure
for i in range(10):
    histogram_plot(i)
```
