## Helper Functions

```python
def apply_model_function(analysis_config, model_function, kwargs={}):
    """
        Takes a function of the form "func(model)" and applies it to each model in 
        analysis_config
    """
    out_dict = {}
    
    for type_name in analysis_config.keys():
        out_dict[type_name] = []
        for path in analysis_config[type_name]["wandb_paths"]:
            # load model
            model = load_model(path)
            # apply model
            function_output = model_function(model, **kwargs)
            # add outcome to dict
            out_dict[type_name].append(function_output)
    
    return out_dict

def triplet_loss_from_name(name):
    if name is "VAETripletLoss":
        return VAETripletLoss()
    elif name is "MaskedVAETripletLoss":
        return MaskedVAETripletLoss(masks=False, device="cpu")
    elif name is "IsolatedTripletLoss":
        return IsolatedTripletLoss()
    else:
        raise Exception(f"Triplet loss : {name} is not recognzied")
        
def average_localizer_metrics(metrics):
    out_metrics = {key: [] for key in metrics[0].keys()}
    for metric in metrics:
        for metric_key in metric.keys():
            out_metrics[metric_key].append(metric[metric_key])
    averaged_metrics = {}
    variance_metrics = {}
    for key in out_metrics.keys():
        lists = out_metrics[key]
        averaged_metrics[key] = np.mean(np.array(lists), axis=0)
        variance_metrics[key] = np.var(np.array(lists), axis=0)
    
    return averaged_metrics, variance_metrics

def triplet_forward(model, triplet_data):
    anchor, positive, negative, _ = triplet_data
    anchor_mean, anchor_logvar, _, _ = model.forward(anchor)
    anchor_mean, anchor_logvar = anchor_mean.unsqueeze(0), anchor_logvar.unsqueeze(0)
    positive_mean, positive_logvar, _, _ = model.forward(positive)
    positive_mean, positive_logvar = positive_mean.unsqueeze(0), positive_logvar.unsqueeze(0)
    negative_mean, negative_logvar, _, _ = model.forward(negative)
    negative_mean, negative_logvar = negative_mean.unsqueeze(0), negative_logvar.unsqueeze(0)

    return (anchor_mean, anchor_logvar), (positive_mean, positive_logvar), (negative_mean, negative_logvar)
    
def apply_loss_function(analysis_config, loss_function, wanted_output="triplet_loss", num_examples=500):
    """
        Calculates the triplet loss for a given model
    """
    def calculate_model_triplet_loss(model, loss_function=None):
        assert not triplet_dataset is None
        # setup the loss
        losses = []
        if loss_function is None:
            loss_function = model.loss_function
            print(loss_function)
            if isinstance(loss_function, MaskedVAETripletLoss):
                loss_function.device = "cpu"
                
        for example_index in range(num_examples):
            triplet_data, _ = triplet_dataset[example_index]
            triplet_data = triplet_forward(model, triplet_data)
            # make dummy data for the loss
            mean, logvar, _, recon = model.forward(image_dataset[example_index])
            real_data, fake_data, mean, logvar, triplet_data = image_dataset[example_index], recon, mean, logvar, triplet_data
            # calculate the loss
            loss_dict = loss_function(real_data, fake_data, mean, logvar, triplet_data)
            loss = loss_dict[wanted_output].item()
            # add it to triplet_losses
            losses.append(loss)
        loss = np.mean(losses)
        return loss
    # get analysis datasets
    image_dataset, metadata_dataset, triplet_dataset = get_analysis_datasets(analysis_config)
    # apply the loss function
    loss_dict = apply_model_function(analysis_config,
        calculate_model_triplet_loss,
        kwargs={
            "loss_function": loss_function
        }
    )
    return loss_dict

def load_experiment_config(wandb_path):
    # load up the experiment parameters from the given run_name
    run_directory = os.path.join(os.environ["LATENT_PATH"], "auto_localization", "logs", wandb_path)
    # load the expeirment config
    params_path = run_directory + "/params.pkl"
    with open(params_path, "rb") as f:
        experiment_config = pickle.load(f)
    
    return experiment_config

def load_model(wandb_path):
    # get the directory
    run_directory = os.path.join(os.environ["LATENT_PATH"], "auto_localization", "logs", wandb_path)
    # get the model config
    experiment_config = load_experiment_config(wandb_path)
    model_config = experiment_config["model_config"]
    # get the model path
    model_weight_path = os.path.join(run_directory, "best_model.pkl")
    # use the model factory to load the model
    model = model_factory.get_model_from_config(model_config["model_type"], model_config)
    # load the model weights
    model.load_state_dict(torch.load(model_weight_path))
    model.eval()
    return model

def load_localizers(wandb_path):
    """
        Gives the localizers and localizer_metrics for a given wandb_path
    """
    # get the directory
    run_directory = os.path.join(os.environ["LATENT_PATH"], "auto_localization", "logs", wandb_path)
    # load a single set of localizers
    localizers_path = os.path.join(run_directory, "localizers.pkl")
    localizers = None
    #with open(localizers_path, "rb") as f:
    #    localizers = pickle.load(f)
    # metrics path
    metrics_path = os.path.join(run_directory, "localizer_metrics.pkl")
    with open(metrics_path, "rb") as f:
        metrics = pickle.load(f)
    
    return localizers, metrics

def apply_localizer_metrics_function(analysis_config, metric_function):
    out_dict = {}
    for model_name in analysis_config.keys():
        out_dict[model_name] = []
        localizer_paths = analysis_config[model_name]["localizer_paths"]
        localizer_metrics_lists = [load_localizers(localizer_path) for localizer_path in localizer_paths]
        metrics = [metrics[1] for metrics in localizer_metrics_lists]
        all_metrics = []
        for loc_metrics in metrics:
            all_metrics.extend(loc_metrics)
        
        func_out = metric_function(all_metrics)
        out_dict[model_name] = func_out
        
    return out_dict

def load_models(analysis_config):
    return apply_model_function(analysis_config, load_model)

def load_analysis_localizers(analysis_config):
    localizers_dict = {}
    metrics_dict = {}
    for type_name in analysis_config.keys():
        localizers_dict[type_name] = []
        metrics_dict[type_name] = []
        for path in analysis_config[type_name]["wandb_paths"]:
            localizers, metrics = load_localizers(path)
            localizers_dict[type_name].append(localizers)
            metrics_dict[type_name].append(metrics)
            
    return localizers_dict, metrics_dict

def get_analysis_datasets(analysis_config, train=False):
    """
        ASSUMPTION: Each model in the analysis config uses the same data
        as input
    """
    def load_datasets_from_wandb(wandb_path):
        # get experiment config
        experiment_config = load_experiment_config(wandb_path)
        # make the datasets from the config
        which_digits = experiment_config["dataset_config"]["which_digits"]
        component_weighting = experiment_config["dataset_config"]["component_weighting"]
        one_two_ratio = experiment_config["dataset_config"]["one_two_ratio"]
        attribute_return = experiment_config["dataset_config"]["attribute_return"]
        single_feature_triplet = experiment_config["dataset_config"]["single_feature_triplet"]
        indexed = experiment_config["dataset_config"]["indexed"]
        print(train)
        # make metadata oracle as trian oracle
        metadata_dataset = MetadataDataset(
            train=False,
            which_digits=which_digits
        )
        metadata_oracle = IndexedMetadataOracle(
            metadata_dataset=metadata_dataset,
            component_weighting=component_weighting,
        )
        # setup data manager
        triplet_dataset = TripletDataset(
            train=train,
            which_digits=which_digits,
            oracle=metadata_oracle,
            one_two_ratio=one_two_ratio,
            attribute_return=attribute_return,
            single_feature_triplet=single_feature_triplet,
        )
        image_dataset = ImageDataset(
            train=train,
            which_digits=which_digits
        )
            
        return image_dataset, metadata_dataset, triplet_dataset

    wandb_path = analysis_config[list(analysis_config.keys())[0]]["wandb_paths"][0]
    image_dataset, metadata_dataset, triplet_dataset = load_datasets_from_wandb(wandb_path)
    
    return image_dataset, metadata_dataset, triplet_dataset
```

### Triplet Loss Experiment

```python
def triplet_loss_comparison_plot(analysis_config):
    # calculate triplet loss for each item in the analysis
    loss_function = IsolatedTripletLoss()
    triplet_loss_dict = apply_loss_function(analysis_config, loss_function, wanted_output="triplet_percentage", num_examples=500)
    # plot the triplet losses
    num_models = len(list(triplet_loss_dict.keys()))
    scale_factor = 2
    fig, axs = plt.subplots(1, 1, figsize=(0.75*scale_factor, num_models*scale_factor))
    model_names = list(triplet_loss_dict.keys())
    triplet_losses = [triplet_loss_dict[model_name] for model_name in list(triplet_loss_dict.keys())]
    plt.title("Triplet Losses")
    plt.violinplot(triplet_losses, vert=True)
    plt.xticks(range(1, len(model_names) + 1), model_names)
    plt.show()
```

## Localization Metadata Plot

```python
def metadata_error_plot(analysis_config):
    def get_last(arrs, loss):
        lasts = []
        for arr in arrs:
            lasts.append(arr[loss][-1])
        return lasts
    metadata_error_dict = apply_localizer_metrics_function(analysis_config, lambda x: get_last(x, "metadata_loss"))
    # plot the triplet losses
    num_models = len(list(metadata_error_dict.keys()))
    scale_factor = 2
    fig, axs = plt.subplots(1, 1, figsize=(num_models*scale_factor, 0.5*scale_factor))
    model_names = list(metadata_error_dict.keys())
    metadata_losses = [metadata_error_dict[model_name] for model_name in list(metadata_error_dict.keys())]
    plt.title("Final Posterior Estimate Metadata Error")
    plt.violinplot(metadata_losses, vert=True)
    plt.yticks(range(1, len(model_names) + 1), model_names)
    plt.show()
```

## Localization Over Queries

```python
import matplotlib
hfont = {'fontname':'Helvetica'}
matplotlib.style.use("seaborn")
matplotlib.rcParams['axes.prop_cycle'] = matplotlib.cycler(color=["r", "b", "g", "purple"])
sns.set_theme()

def plot_localization(analysis_config, title="Title", filename="filename", legend=True, axs=None):
    localizers_dict, metrics_dict = load_analysis_localizers(analysis_config)
    loss_name = "nearest_neighbor_loss"
    dictionary = apply_localizer_metrics_function(analysis_config, average_localizer_metrics)
    # convert to 
    for model_type in dictionary:
        mean, variance = dictionary[model_type]
        mean = mean[loss_name]
        variance = variance[loss_name]
        # make the lines
        axs.plot(mean, label=model_type)
        axs.fill_between(np.arange(len(mean)), mean-variance, mean+variance, alpha=0.2)

    axs.set_xlabel("Number of Queries", **hfont, fontsize=14)
    axs.set_ylim(0, 1.25)
    axs.set_xlim(0, 29)
    if legend:
        axs.set_ylabel("Nearest Neighbor \n Metadata Loss", **hfont, fontsize=14)
        axs.legend()
    axs.set_title(title, fontsize=14)
```

```python
def plot_localization_over_queries(localizer_metrics):
    # unpack the metrics
    means, variances = localizer_metrics
    metadata_losses = means["metadata_loss"]
    localization_losses = means["localization_loss"]
    metadata_variance = variances["metadata_loss"]
    localization_variance = variances["localization_loss"]
    latent_covariance = means["latent_covariance_determinant"]
    latent_covariance = np.concatenate((np.array([latent_covariance[0]]), latent_covariance))
    # make plots for these losses
    fig, ax = plt.subplots(1, 2, figsize=(10,5))
    metadata_train = np.arange(0, np.shape(metadata_losses)[0])
    ax[0].set_title("Metadata Loss Plot")
    ax[0].plot(metadata_train, metadata_losses, label="Metadata Loss")
    ax[0].fill_between(metadata_train, metadata_losses-metadata_variance, metadata_losses+metadata_variance, alpha=0.2)
    ax[0].set_ylim(0)
    # plot test on the right axis
    localization_train = np.arange(0, np.shape(localization_losses)[0])
    ax[1].set_title("Latent Loss Plot")
    ax[1].plot(localization_train, localization_losses, label="Latent Loss")
    ax[1].set_ylabel("Latent Loss")
    ax[1].fill_between(localization_train, localization_losses-localization_variance, localization_losses+localization_variance, alpha=0.2)
    # plot secondary axis on the right
    right_ax = ax[1].twinx()
    # make a plot with different y-axis using second axis object
    right_ax.plot(localization_train, latent_covariance, color="orange", label="Latent Covariance Determinant")
    right_ax.set_ylabel("Latent Covariance Determinant", color="orange", fontsize=14)
    # axis labels/title
    plt.xlabel("Num Queries")
    plt.legend(loc="upper left")
    plt.show()
```
