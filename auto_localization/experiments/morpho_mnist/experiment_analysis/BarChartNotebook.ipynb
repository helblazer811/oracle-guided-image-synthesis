```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def plot_dictionary(data, color, ax, title=""):
    plt.rcdefaults()
    plt.rc('font', family='Helvetica')
    # Example data
    model_names = list(data.keys())
    y_pos = np.arange(len(model_names))
    data_values = list(data.values())

    ax.barh(y_pos, data_values, align='center', color=color, edgecolor="black", linewidth=1.0)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.set_yticks(y_pos)
    ax.set_yticklabels(model_names)
    for i, v in enumerate(data_values):
        ax.text(v*1.05, i + 0.1, "{:3}".format(v), color=color, fontweight='bold')
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel(title)

```

```python
# data
triplet_percentiles = {
    "No Triplet Loss":  0.741*100,
    "Traditional Triplet Loss": 0.996*100,
    "Bayesian Triplet Loss": 0.995*100
}

reconstruction_error = {
    "No Triplet Loss": 0.0358,
    "Traditional Triplet Loss": 0.0355,
    "Bayesian Triplet Loss": 0.0370
}
```

```python
fig, axs = plt.subplots(2, 1, figsize=(4, 3.0))
print(axs)
plot_dictionary(triplet_percentiles, "mediumblue", axs[0], title='Percentage of Triplets Satisfied (%)')
plot_dictionary(reconstruction_error, "darkorange", axs[1], title="Model Reconstruction Error")
plt.subplots_adjust(hspace=0.6)
plt.savefig("percentile_bar_graphs.pdf", dpi=500, figsize=(5, 3.0), bbox_inches='tight')
plt.show()
```
